```python
feats = pd.concat((df1,df2))
feats = pd.concat((feats,df3))
feats = pd.concat((feats,df4))
feats = pd.concat((feats,df5))
feats = pd.concat((feats,df6))
feats = pd.concat((feats,df7))
feats = pd.concat((feats,df8))
feats = pd.concat((feats,df9))
feats = pd.concat((feats,df10))
feats = pd.concat((feats,df11))
feats = pd.concat((feats,df12))
feats = pd.concat((feats,df13))
feats = pd.concat((feats,df14))
feats = pd.concat((feats,df15))
feats = pd.concat((feats,df16))
feats = pd.concat((feats,df17))

feats = feats.reset_index(drop=True)
```

```python
features = feats.groupby('ID')['认定类型'].count().reset_index(drop=False)
features.rename(columns= {'认定类型':'科技奖项拼接个数'},inplace=True)
type_class = feats['认定类型'].unique()
type_class = type_class[~pd.isna(type_class)]
jibie_class = feats['级别'].unique() 
jibie_class = jibie_class[~pd.isna(jibie_class)]
for year in range(2018,2024):
    sub_data =feats[feats['年度']==year].groupby('ID')['认定类型'].count().reset_index(drop=False)
    sub_data.rename(columns= {'认定类型':f'{year}年_政府支持拼接个数'},inplace=True)
    features = features.merge(sub_data,how='left',on='ID')
    for cl in type_class:
        sub_data = feats[(feats['认定类型']== cl)&(feats['年度']==year)].groupby('ID')['级别'].count().reset_index(drop=False)
        sub_data.rename(columns={'级别':f'{year}年_政府支持拼接_{cl}_数量'},inplace=True)
        features = features.merge(sub_data,how='left',on='ID')
    for cl in jibie_class:
        sub_data = feats[(feats['级别']== cl)&(feats['年度']==year)].groupby('ID')['认定类型'].count().reset_index(drop=False)
        sub_data.rename(columns={'认定类型':f'{year}年_政府支持拼接_{cl}_数量'},inplace=True)
        features = features.merge(sub_data,how='left',on='ID')     
#df = df.merge(feature_yanfashili,how='left',on='ID')
```

```python
data_all = data_all.merge(features,how='left',on='ID')
```
